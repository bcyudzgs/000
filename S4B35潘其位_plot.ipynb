```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)
def plotsine(A = 1, omega = 1, phi = 1):
    y = A * np.sin(omega * x + phi)
    plt.plot(x, y)
def plotcosine(A = 1, omega = 1, phi = 1):
    y = A * np.cos(omega * x + phi)
    plt.plot(x, y)
plotsine()
plotcosine()
```
