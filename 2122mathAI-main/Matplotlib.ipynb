# Let's try $f(x) = \dfrac{\sin (5x)}{1 + x^2}$

```python
x = np.linspace(0, 10, 100)
y = np.sin(5*x)/(1 + x**2)
plt.plot(x, y)
```

# Try something special

```python
x = np.linspace(0, 10, 100)
y = np.sin(5*x)/(1 + x**2)
plt.plot(x, y, marker = 'o')
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(5*x)/(1 + x**2)
plt.plot(x, y, 'b:')
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.plot(x[y>0], y[y>0], 'ro')
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
plt.plot(x, y, lw=3)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
          ['$0$', '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$']);
```

```python
x = np.linspace(0, 2*np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.plot(x, y_sin, label = "$\sin$")
plt.plot(x, y_cos, label = "$\cos$")
plt.legend(loc = 2)
```

```python
def draws():
    x = np.linspace(0, 2*np.pi, 100)
    y_sin = np.sin(x)
    y_cos = np.cos(x)
    plt.plot(x, y_sin, label = "$\sin$")
    plt.plot(x, y_cos, label = "$\cos$")
    plt.legend()
    plt.grid(color='k', linestyle='--', alpha = 0.3, linewidth=1)
```

```python
draws()
```
