```python
import cv2
import time
import math
import numpy as np
cap = cv2.VideoCapture(0)
ratio = cap.get(cv2.CAP_PROP_FRAME_WIDTH) / cap.get(cv2.CAP_PROP_FRAME_HEIGHT)# Get the ratio of camera
WIDTH = 600
HEIGHT = int(WIDTH/ratio)
num = 3
now = 0
size = 1
count = 0
start = False

cv2.namedWindow('me',cv2.WINDOW_NORMAL)

ret = True
while ret:
    ret, frame = cap.read()
    frame = cv2.resize(frame, (WIDTH, HEIGHT))
    frame = cv2.flip(frame, 1)
    if cv2.waitKey(1) == ord(' '):
        now = math.ceil(time.time())
        start =True
    if start:
        cv2.putText(frame, str(num), (260,290), cv2.FONT_HERSHEY_SIMPLEX, size, (0,0,0), size, cv2.LINE_AA)
    if start and round(time.time(),3) - now == 0.125:
        if count < 5:
            size += 1
            count +=1
        else:
            size -= 1
    if start and math.floor(time.time()) - now == 1:
        now = math.ceil(time.time())
        num -= 1
        count = 0
    if start and num == 0:
        ret, frame = cap.read()
        frame = cv2.resize(frame, (WIDTH, HEIGHT))
        frame = cv2.flip(frame, 1)
        cv2.imwrite("image.jpeg", frame)
        frame = np.zeros((600,600,3), dtype = np.uint8)
        frame.fill(255)
        num = 3
        start = False
    
    cv2.imshow('me', frame)
    
    if cv2.waitKey(1) == 27:
        break
        
cv2.destroyAllWindows()
cap.release()
```
