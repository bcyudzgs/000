```python
from openpyxl import Workbook, load_workbook 
from openpyxl.utils import get_column_letter
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
wb = load_workbook("E_DEM_FR_2021_Q2.xlsx")
ws = wb.active
x=[0,1,2,3,4,5,6]
ym=[0]*7
yf=[0]*7
for i in range(7):
    ym[i]=ws["D"+str(i*2+9)].value
    yf[i]=ws["D"+str(i*2+10)].value
plt.xticks(list(np.arange(0,7,1)),
                  ['< 25','25 - 29','30 - 34','35 - 39','40 - 44','45 - 54',' 55']);
plt.plot(x, ym, label = "Male")
plt.plot(x, yf, label = "Female")
plt.legend(loc = 2)
```
